```python
import json
import imgkit
options = {'encoding': "UTF-8"}

# path to the file with XML tables
table_path = '../data/en/tables/tables_xml.json'
# paths to store (1) xml versions of the tables, (2) images of the tables
xml_loc = '../data/en/tables/xml/'
triples_loc = '../data/en/tables/triples/'

with open(table_path, 'r') as f:
    triples = json.load(f)
```

```python
table_style = '''

<style type="text/css">
.tg  {border-collapse:collapse;border-spacing:0;}
.tg td{border-color:black;border-style:solid;border-width:1px;font-family:Arial, sans-serif;font-size:14px;
  overflow:hidden;padding:10px 5px;word-break:normal;}
.tg th{border-color:black;border-style:solid;border-width:1px;font-family:Arial, sans-serif;font-size:14px;
  font-weight:normal;overflow:hidden;padding:10px 5px;word-break:normal;}
.tg .tg-0pky{border-color:inherit;text-align:left;vertical-align:top}
.tg .tg-0lax{text-align:left;vertical-align:top}
    
.center {
  margin: 0;
  position: absolute;
  left: 5%;
  -ms-transform: translate(-50%, -50%);
  transform: translate(-50%, -50%);
}

</style>

'''
```

```python
# compile images and save them
for item_id, tr in triples.items():
    tr = tr.replace('\n', '')
    full_xml = table_style + tr
    with open(f'{xml_loc}{item_id}.html', 'w') as f1:
        f1.write(full_xml)
    imgkit.from_file(f'{xml_loc}{item_id}.html', f'{triples_loc}{item_id}.jpg', options=options)
```
